Handling EXCEL

Read from an EXCEL file

```python
import pandas as pd

df = pd.read_excel('jaish.xlsx')
print(df)
```

```python
#Displays the statistics
print(df.describe())
```

```python
#Displays the rows and columns
print(df.shape)
```

```python
#Statistical Information about Excel file
print(df.info())
```

Extract the contents

```python
#Extract all rows where Units Sold > 500
high_sales = df[df['Units Sold'] > 500]
print("\nCars with Units Sold > 500:")
print(high_sales)
```

```python
#Extract only BMW 
bmw_cars = df[df['Car Brand'] == 'BMW']
print("\nBMW Cars:")
print(bmw_cars)
```

```python
#Select only Car Brand and Total Sales columns
sales_summary = df[['Car Brand', 'Total Sales ($)']]
print("\nCar Brand with Total Sales:")
print(sales_summary)
```

Write the output of a Query into an EXCEL File

```python
# Write the query results to a new Excel file

#Extract cars with Units Sold > 500
high_sales_df = df[df['Units Sold'] > 500]
print("\nCars with Units Sold > 500:")
print(high_sales_df)

#Write the query results to a new Excel file
output_file = '/mnt/data/high_sales.xlsx'
high_sales_df.to_excel(output_file, index=False)
```

Read a Excel Chunk-by-chunk

```python
import pandas as pd

#Convert Excel to CSV
df = pd.read_excel("jaish.xlsx")
df.to_csv("jaish.csv", index=False)

# read the CSV in chunks
for i, chunk in enumerate(pd.read_csv("jaish.csv", chunksize=10)):
    print(f"\n--- Chunk {i+1} ---")
    display(chunk)  
```

Writing numeric data to a new Excel file

```python
# Create numeric data
data = {
    "Product_ID": [101, 102, 103, 104],
    "Quantity_Sold": [20, 35, 15, 50],
    "Price_Per_Unit": [100, 150, 120, 80],
    "Total_Sales": [20*100, 35*150, 15*120, 50*80]  # Calculated column
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Write to Excel
output_file = "numeric_data_ExcelSheet.xlsx"
df.to_excel(output_file, index=False)

print(f"Numeric data written successfully to {output_file}")
```

```python
data = pd.read_excel('numeric_data_ExcelSheet.xlsx')
print(data) # print the numeric data from excel
```

Write text data into a Excel File

```python
import pandas as pd

# Create text data
data = {
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "City": ["New York", "Los Angeles", "Chicago", "Houston"],
    "Department": ["HR", "IT", "Finance", "Marketing"]
}

#Convert to DataFrame
df = pd.DataFrame(data)

#Write to Excel
output_file = "text_data_excelsheet.xlsx"
df.to_excel(output_file, index=False)

print(f"Text data written successfully to {output_file}")
```

```python
data = pd.read_excel('text_data_excelsheet.xlsx')
print(data) # print the text data from excel
```
