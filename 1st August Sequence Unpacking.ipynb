Unpacking Sequence

```python
#List of odd numbers
ODDS = [1, 3, 5, 7]
```

```python
#Unpacking the list into variables
a, b, c, d = ODDS
```

```python
#Outputs
print(a)
print(b) 
print(c) 
print(d)
```

```python
#Unpacking with the rest of the elements stored in 'c'
a, b, *c = ODDS
```

```python
print(a)
print(b)  
print(c)
```

```python
#Unpacking with first and last elements stored separately, and the middle in 'b'
a, *b, c = ODDS
```

```python
print(a) 
print(b)  
print(c)
```

```python
#Function to add three numbers
def add_three_numbers(x, y, z):
    return x + y + z
```

```python
print(add_three_numbers(*ODDS[:3]))
```

```python
#Function to add four numbers
def add_four_numbers(w, x, y, z):
    return w + x + y + z
```

```python
print(add_four_numbers(*ODDS))
```

```python
#Function demonstrating unpacking with variable number of string arguments
def greet_all(*names):
    for name in names:
        print(f"Hello, {name}!")
```

```python
greet_all("Jaish", "Keshav", "Ananya", "Jyo")
```

```python
#Function demonstrating unpacking dictionary into keyword arguments
def display_info(**details):
    for key, value in details.items():
        print(f"{key} : {value}")
```

```python
info_data = {
    "Code": "Advanced Python Programming",
    "School": "SCOPE",
    "Course": "CSI4007"
}

display_info(**info_data)
```
