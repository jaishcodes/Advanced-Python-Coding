Image Tutorial with the help of matplotlib documentation

```python
#Importing the image from the python image library
from PIL import Image

#Import the pyplot module from matplotlib
#Used for plotting and displaying images/graphs.
import matplotlib.pyplot as plt

#Import numpy for numerical operations
import numpy as np
```

```python
#Open the image and then convert it into Numpy array this transforms it inot a matix of pixel values
img = np.asarray(Image.open('Beetle.jpg'))
#Prints the representation of the Numpy array repr() shows all the values and data types
print(repr(img))
```

```python
#Displays the image using the imshow() function
#Also renders pixel values from Numpy array as an image
imgplot = plt.imshow(img)
```

```python
#Extracting only the first channel of the image 

#: selects all rows 
#: selects all columns 
#0 selects the first channel that is red
lum_img = img[:, :, 0]
#Display the red channel image using matplotlib
#This will show only intensity values of the red channel
plt.imshow(lum_img)
```

```python
#Display the red channel but applying the hot colormap
#hot colormap maps low values to black, mid values to red/orange and high values to yellow/white
#This creats a heatmap effect
plt.imshow(lum_img, cmap="hot")
```

```python
#Read and display red channel image using imshow()
imgplot = plt.imshow(lum_img)
#Changes the colormap to nipy_spectral
#This has a lot more colors
imgplot.set_cmap('nipy_spectral')
```

```python
#Read and display red channel image using imshow()
imgplot = plt.imshow(lum_img)
#Add colorbar next to image
#Colorbar shows mapping between pixel intensity values and their corresponding colors
plt.colorbar()
```

```python
#Flatten the 2D image array into a 1D array using .ravel()
#Single list of all pixel intensity values is given
# Then, plot a histogram of these values using 256 bins 
#fc='k' sets the face color of the bars to black
#ec='k' sets the edge color of the bars to black
plt.hist(lum_img.ravel(), bins=range(256), fc='k', ec='k')
```

```python
#Read and display red channel image using imshow()
#Set the color limits with clim=(0, 175)
#Pixels with values <= 0 will appear as the lowest color that is black
#Pixels with values >= 175 will appear as the highest color 
#Values in between are scaled linearly
plt.imshow(lum_img, clim=(0, 175))
```

```python
#Read and display red channel image using imshow()
imgplot = plt.imshow(lum_img)
#Set the color limits of the image object
#Pixels <= 0 → minimum color, pixels >= 175 → maximum color
imgplot.set_clim(0, 175)
```

```python
#Open the image using PIL
img = Image.open('Beetle.jpg')
# Resize the image in place to fit within a 64x64 pixel box
img.thumbnail((64, 64))
#Display the resized image using matplotlib
imgplot = plt.imshow(img)
```

```python
#Display the image using matplotlib's imshow()
#Set interpolation to bilinear to get smoother images
#bilinear interpolation calculates pixel values using linear interpolation between neighboring pixels
imgplot = plt.imshow(img, interpolation="bilinear")
```

and bicubic:



```python
#Display the image using matplotlib's imshow()
#Set interpolation to bicubic to get smoother images
#bicubic interpolation uses cubic polynomials to calculate pixel values from neighbors
imgplot = plt.imshow(img, interpolation="bicubic")
```

Visualization of Different Types of Data
    -Numerical 
    -Categorical 
    -Text 
    -Ordinal

```python
import matplotlib.pyplot as plt
```

```python
#Numerical Data
numerical = [12, 15, 20, 18, 25, 30, 22, 18]
plt.figure(figsize=(10,6))
plt.hist(numerical, bins=5, color='red',edgecolor='black')
plt.title("Numerical Data - Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
#Categorical Data
categories = ['Ford', 'Volkswagen', 'Ferrari', 'Porsche']
counts = [28, 25, 15, 18]
plt.figure(figsize=(10,6))
plt.bar(categories, counts, color='red', edgecolor='black')
plt.title("Categorical Data - Bar Chart")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()
```

```python
#Text Data
words = ['Ford Mustang GTD GT3', 'Corvette C8.R', 'Ferrari 296 GT3', 'Porsche 911 GT3']
freq = [8, 6, 4, 14]
plt.figure(figsize=(10,6))
plt.bar(words, freq, color='red', edgecolor='black')
plt.title("Text Data - Word Frequency")
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.show()
```

```python
#Ordinal Data
grades = ['Poor', 'Average', 'Good', 'Excellent']
scores = [20, 50, 65, 80]  
plt.figure(figsize=(10,6))
plt.bar(grades, scores, color='red', edgecolor='black')
plt.title("Ordinal Data - Ordered Bar Chart")
plt.xlabel("Grade")
plt.ylabel("Score")
plt.show()
```
