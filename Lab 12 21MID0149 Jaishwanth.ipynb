Read a CSV Chunk by chunk

```python
chunk_size = 100
count = 6

for chunk in pd.read_csv("customer_data.csv", chunksize=chunk_size): 
    print(f"DataFrame Chunk {count}:") 
    print(chunk)
    count += 1
```

Append to a CSV

```python
import csv

new_data = [
    ['Email', 'PurchaseDate', 'PurchaseAmount', 'Product'],
    ['Jaish', 'July 1, 2025', 1300, 'Laptop'],
    ['Jyo', 'July 5, 2025', 900, 'Tablet']
]

with open('customer_data.csv', mode='a', newline='') as file:
    writer = csv.writer(file)
    for row in new_data[1:]: 
        writer.writerow(row)

df_appended = pd.read_csv("customer_data.csv")
print(df_appended)
```

Writing to a CSV File

Writing numeric data to a new CSV file

```python
from csv import writer

new_row = ['Keshav', 'July 1, 2025', 1300, 'Laptop']

with open('customer_data.csv', 'a', newline='') as file:
    csv_writer = writer(file)
    csv_writer.writerow(new_row)

df_appended = pd.read_csv("customer_data.csv")
print(df_appended)
```

Write text data into a CSV File

```python
import csv

data = [
    ['Email', 'PurchaseDate', 'PurchaseAmount', 'Product'],
    ['Jyo', 'January 6, 2025', 1250, 'Smartphone'],
    ['Jaish', 'January 8, 2025', 875, 'Laptop'],
    ['Bhav', 'January 31, 2025', 2000, 'Smartphone']
]

with open('customer_data.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(data)

df_appended = pd.read_csv("customer_data.csv")
print(df_appended)
```
